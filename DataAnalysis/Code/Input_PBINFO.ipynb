```python
import pymysql
import pandas as pd
```

```python
PB_INFO = pd.read_csv('../../DataBase/DataBase_Data/PB_INFO.csv', encoding = 'euc-kr')
print(PB_INFO.columns)
PB_INFO.head()
```

```python
conn = pymysql.connect(
    host='project-db-campus.smhrd.com',
    port=3312,
    user='sc_24K_bigdata31_p3_2',
    password='smhrd2',
    db='sc_24K_bigdata31_p3_2',
    charset='utf8'
)
```

```python
# 커서 생성
cursor = conn.cursor()

# INSERT 쿼리 (pb_idx는 AUTO_INCREMENT이므로 제외)
insert_query = """
INSERT INTO pb_info (
    pb_idx, li_idx, ex_idx, pb_num, pb_ques, pb_detail, 
    pb_choi1, pb_choi2, pb_choi3, pb_choi4, 
    pb_ans, pb_solu, pb_cr, topic_idx
) VALUES (%d, %d, %d, %d, %s, %s, %s, %s, %s, %s, %d, %s, %s, %d)
"""
```

```python
# 데이터 삽입
try:
    for index, row in PB_INFO.iterrows():
        try:
            # CSV 열 이름(대문자)을 소문자 DB 열 이름에 맞게 매핑
            data = (
                row['PB_IDX'], row['LI_IDX'], row['EX_IDX'], row['PB_NUM'], row['PB_QUES'], 
                row['PB_DETAIL'], row['PB_CHOI1'], row['PB_CHOI2'], 
                row['PB_CHOI3'], row['PB_CHOI4'], row['PB_ANS'], 
                row['PB_SOLU'], row['PB_CR'], row['TOPIC_IDX']
            )
            # INSERT 실행
            cursor.execute(insert_query, data)
            print(f"Row {index + 1} inserted successfully.")
        except:
            print("error")
            break

    # 변경 사항 커밋
    conn.commit()

except Exception as e:
    print(f"An error occurred: {e}")
    conn.rollback()

finally:
    # 커서와 연결 닫기
    cursor.close()
    conn.close()
```
